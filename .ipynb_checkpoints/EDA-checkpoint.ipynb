```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from ucimlrepo import fetch_ucirepo

# Fetch dataset from UCI ML Repository (ID 296)
diabetes_data = fetch_ucirepo(id=296)
X = diabetes_data.data.features
y = diabetes_data.data.targets

# Make sure target column has a nice name
if "readmitted" not in y.columns:
    y.columns = ["readmitted"]

# Combine into a single DataFrame
df = pd.concat([X, y], axis=1)

print("Shape:", df.shape)
print(df.head())
print(df.columns)
print(y.head())
```

```python
# dataset table summarising columns (dtype, uniques, missing %)
col_summary = pd.DataFrame({
    "dtype": df.dtypes.astype(str),
    "n_unique": df.nunique(),
    "missing_pct": df.isna().mean().round(3)
}).sort_index()

print("\nColumn summary (first 25 rows):")
display(col_summary.head(25))
col_summary.to_csv("column_summary_diabetes.csv", index=True)

# missing values per column (top 10) + barplot
print("\nMissing values per column (top 10):")
print(df.isna().sum().sort_values(ascending=False).head(10))

missing_percent = df.isna().mean().sort_values(ascending=False) * 100
plt.figure(figsize=(10, 4))
sns.barplot(x=missing_percent.head(10), y=missing_percent.head(10).index)
plt.title("Top 10 Columns with Missing Values (%)")
plt.xlabel("Percentage Missing")
plt.ylabel("Feature")
plt.show()

# duplicates
print("\nNumber of duplicate rows:", df.duplicated().sum())
```

```python
# List of key numeric columns 
numeric_cols = [
    "time_in_hospital",
    "num_lab_procedures",
    "num_procedures",
    "num_medications",
    "number_outpatient",
    "number_emergency",
    "number_inpatient",
    "number_diagnoses",
]

# numeric summary table for key columns
summary_stats = df[numeric_cols].describe().T
print("\nNumeric summary for key features:")
print(summary_stats)
summary_stats.to_csv("numeric_summary_diabetes.csv")

# outlier count in num_lab_procedures
feature = "num_lab_procedures"
Q1, Q3 = df[feature].quantile([0.25, 0.75])
IQR = Q3 - Q1
outliers = ((df[feature] < (Q1 - 1.5 * IQR)) | (df[feature] > (Q3 + 1.5 * IQR))).sum()
print(f"\nOutliers detected in '{feature}': {outliers}")

# outlier visualization
plt.figure(figsize=(4, 3))
plt.boxplot(df[feature].dropna(), vert=True, showfliers=True)
plt.title(f"Boxplot of {feature}")
plt.ylabel(feature)
plt.tight_layout()
plt.show()
```

```python
# Select all numeric columns for correlation heatmap
all_numeric_cols = df.select_dtypes(include=["int64", "float64"]).columns

# Distribution of hospital stay duration
plt.figure(figsize=(8, 4))
sns.histplot(df["time_in_hospital"], bins=15, kde=True)
plt.title("Distribution of Hospital Stay Duration")
plt.xlabel("Days in Hospital")
plt.ylabel("Count")
plt.show()

# Correlation heatmap of numeric features
plt.figure(figsize=(10, 6))
corr = df[all_numeric_cols].corr()
sns.heatmap(corr, cmap="coolwarm", annot=False)
plt.title("Correlation Heatmap of Numeric Features")
plt.tight_layout()
plt.show()
```

```python
# binary label 1=<30 days, 0 otherwise
df["readmit_30d"] = (df["readmitted"] == "<30").astype(int)

print("\nOriginal readmitted value counts:")
print(df["readmitted"].value_counts())

print("\nBinary readmit_30d value counts:")
print(df["readmit_30d"].value_counts())

# class distribution plot (3-class)
plt.figure(figsize=(4, 3))
df["readmitted"].value_counts(normalize=True).sort_index().plot(kind="bar")
plt.ylabel("Proportion of encounters")
plt.title("Distribution of Readmission Categories")
plt.tight_layout()
plt.show()
```

```python
# Early readmission by age band
age_readmit = (
    df.groupby("age")["readmit_30d"]
      .mean()
      .sort_index()
)

plt.figure(figsize=(6, 3))
age_readmit.plot(kind="bar")
plt.ylabel("P(readmit < 30 days)")
plt.title("Early Readmission Rate by Age Band")
plt.tight_layout()
plt.show()

# Early readmission by race
plt.figure(figsize=(6, 3))
(df.groupby("race")["readmit_30d"].mean()
   .sort_values(ascending=False)
   .plot(kind="bar"))
plt.ylabel("P(readmit < 30 days)")
plt.title("Early Readmission Rate by Race")
plt.tight_layout()
plt.show()

# Early readmission by length-of-stay bins
df["los_bin"] = pd.cut(
    df["time_in_hospital"],
    bins=[0, 3, 5, 7, 14],
    labels=["13", "45", "67", "814"]
)

plt.figure(figsize=(4, 3))
(df.groupby("los_bin")["readmit_30d"].mean()
   .plot(kind="bar"))
plt.ylabel("P(readmit < 30 days)")
plt.title("Early Readmission Rate by Length of Stay")
plt.tight_layout()
plt.show()
```

```python
# Make a numeric version of the 3-class readmission label
# NO -> 0, >30 -> 1, <30 -> 2
df["readmitted_flag"] = df["readmitted"].map({"NO": 0, ">30": 1, "<30": 2})

# Get all numeric feature columns
all_numeric_cols = df.select_dtypes(include=["int64", "float64"]).columns

# Correlation of each numeric feature with readmission
baseline_corr = (
    df[all_numeric_cols]
      .corrwith(df["readmitted_flag"])
      .sort_values(ascending=False)
)

print("\nTop correlations with readmitted_flag:")
print(baseline_corr.head(10))
```

```python
# Pick a smaller set of important numeric features
numeric_cols = [
    "time_in_hospital",
    "num_lab_procedures",
    "num_procedures",
    "num_medications",
    "number_outpatient",
    "number_emergency",
    "number_inpatient",
    "number_diagnoses",
]

# Correlation between these predictors
corr_matrix = df[numeric_cols].corr()

# Show as a heatmap
plt.figure(figsize=(6, 5))
sns.heatmap(corr_matrix, cmap="coolwarm", center=0)
plt.title("Correlation Matrix of Key Numeric Predictors")
plt.tight_layout()
plt.show()
```
